```python
import numpy as np
import pandas as pd
import random

DF1 = pd.DataFrame()

for i in range(5):
    m = np.array([[random.randint(0, 10), random.randint(0, 10)],
                 [random.randint(0, 10), random.randint(0, 10)],
                 [random.randint(0, 10), random.randint(0, 10)]])
    print(f'm = {m}')

    DF2 = pd.DataFrame(m)
    print(f'DF2 = \n{DF2}')
    DF1 = DF1.append(DF2)
 
    print(f'DF1 = \n{DF1.reset_index(drop=True)}\n')
    
```
