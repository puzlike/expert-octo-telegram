```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
df = pd.read_excel('prodact_lane.xlsx')

def conv_date(ts):
    str = ts.isoformat()
    return datetime.strptime(str[:10], '%Y-%m-%d')

df.insert(1, '日付', df['日時'].map(conv_date))
display(df)

today = df['日付'].iloc[0]
nextday = today + relativedelta(days= 1)
print(nextday)
```
