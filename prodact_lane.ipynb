```python
import pandas as pd
from datetime import datetime
from dateutil.relativedelta import relativedelta
df = pd.read_excel('prodact_lane.xlsx')
display(df)

day = datetime.strptime(df.iloc[0, 0].isoformat()[:10], '%Y-%m-%d') 
nextday = day + relativedelta(days=1)
sdate = day.isoformat() + ' 23:00:00'
edate = nextday.isoformat() + ' 05:00:00'

df1 = df.query('日時 >=@sdate and 日時 <=@edate').reset_index(drop=True)
display(df1)

sum_lanes = []
l1 = len(list(df1.columns))
l2 = len(list(df1.index))
for j in range(1, l1):
    sum_lane = 0
    for i in range(l2):
        sum_lane += df1.iloc[i, j]
    sum_lanes.append(sum_lane)

for i, sum_lane in enumerate(sum_lanes):
    print(f'生産レーン{i + 1}の合計は{sum_lane}です。')

```
