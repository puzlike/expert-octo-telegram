```python
import pandas as pd
input_data = pd.read_excel('選手名競走車名.xlsx')
input_data
```

```python
import MeCab
tagger = MeCab.Tagger()
all_words = []
parts = ["名詞"]
for n in range(len(input_data)):
    text = input_data["選手名競走車名"].loc[n]
    words = tagger.parse(text).splitlines()
    words_arr = []
    for i in words:
        if i == "EOS" or i == "": continue
        word_tmp = i.split()[0]
        if len(i.split()) >= 4:
            part = i.split()[4].split("-")[0]
            if not (part in parts): continue
            words_arr.append(word_tmp)
    all_words.append(words_arr)
print(all_words)


```

```python
data = []
for words in all_words:
    name = ""
    car_name = ""
    for i in range(2):
        name = words[0] + " " + words[1]
    data.append([name])
output_data = pd.DataFrame(data, columns=['選手名'])
output_data
```

```python
output_data.to_excel('選手名.xlsx', index=False)
```
