```python
import pandas as pd
f1 = pd.read_csv('f1.csv')
f1
```

```python
f2 = pd.read_csv('f2.csv')
f2
```

```python
f3 = pd.read_csv('f3.csv')
f3
```

```python
newf = pd.merge(f1, f2, on="name", how="outer")
newf = pd.merge(newf, f3, on="name", how="outer" )
newf
```

```python
newf.to_csv('newf.csv')
```
