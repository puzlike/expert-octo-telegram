```python
import pandas as pd
time_data = pd.read_excel('選手名_タイム.xlsx')
time_data
```

```python
time_data['順位'] = time_data['タイム'].rank(method='min').astype(int).astype(str) + "位"
time_data
```

```python
point = [
    ['1位', '10点'],
    ['2位', '8点'],
    ['3位', '6点'],
    ['4位', '5点'],
    ['5位', '4点'],
    ['6位', '3点'],
    ['7位', '2点'],
    ['8位', '1点']
]
point = pd.DataFrame(point, columns=['順位', '得点'])
point
```

```python
time_data = pd.merge(time_data, point, on='順位', how='left')
time_data
```
