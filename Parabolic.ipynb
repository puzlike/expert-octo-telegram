```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import math
h = int(input("投げ上げる高さ(m)：H = "))
v = int(input("初速度(m/s)：V＝"))
k = int(input("投げ上げ角度(度)：K＝"))
G = 9.8
k = math.radians(k)
x0 = (2/G) * (v * math.sin(k) + math.sqrt((v*math.sin(k))**2 + 2*G*h))
x = np.linspace(0, x0, 50)
y = v * x * math.sin(k) - 1/2 * G * x**2 + h
fig = plt.figure(figsize=(8, 6))
plt.plot(x, y, 'o')
plt.title("ボールの投げ上げ", fontsize=25)
plt.xlabel("時間（s）", fontsize=18)
plt.ylabel("高さ(m)", fontsize=18)
plt.xlim(0,)
plt.ylim(0,)
plt.show()
```
